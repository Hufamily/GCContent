```python
import numpy as np
import matplotlib.pyplot as plt

species=["Llo","Bfl","Mme","Pma","Cmu","Cel","Hsa","Mmu","Ppa","Cre","Ath", "Mco","Ota","Olu","Cso","Bpr","Tko","Osa","Bsu","Ddi","Afu","Mtu"]
x=[0.414, 0.412, 0.38, 0.366, 0.469, 0.354, 0.409, 0.417, 0.337, 0.641, 0.36, 0.64, 0.594, 0.604, 0.66, 0.481, 0.52, 0.437, 0.435, 0.224, 0.498, 0.656]
y=[0.397, 0.385, 0.348, 0.327, 0.449, 0.264, 0.385, 0.397, 0.288, 000.727, 0.247, 0.748, 0.719, 0.728, 0.754, 0.471, 0.454, 0.432, 0.466, 0.137, 0.435, 0.681]
color=["blue","blue","blue","blue","blue","blue","blue","blue","green","green","green","green","green","green","green","green","red","green","yellow","blue","blue","yellow"]

plt.close('all')
plt.plot([0,1],[0,1], color="black", linestyle="dashed")
for i in range(len(species)):
    plt.scatter(x[i],y[i],color=color[i],label=species[i])
plt.xlabel('GC Content of Species')
plt.ylabel('GC Content of Repeats in Species')
plt.legend()

popt, pcov = np.polyfit(x,y,deg=1,cov=True)
err = np.sqrt(np.diag(pcov))#sqrt of diagonal elements of pcov are error
print(err)
popt=np.round(popt,3)#rounds popt to three digits
m,b = popt
func = np.poly1d(popt)

plt.plot(x,func(x), 'm-')
label = str(m)+'x+'+str(b) #Can use latex in python
plt.text(.05, 0.8,label)
```
