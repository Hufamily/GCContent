```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
column_names = ["Species File", "GCR", "GCNR", "GC"]

# Specify data types for each column
dtypes = {
    "Species File": str,
    "GCNR": float,
    "GCR": float,
    "GC": float
}

# Read the CSV file
df = pd.read_csv("gc_content_results.csv", names=column_names, dtype=dtypes, skiprows=1)
```

```python
x = df["GCNR"]
y = df["GCR"]

plt.scatter(x[0], y[0], color="red", label = "Placozoa")
plt.scatter(x[1:9], y[1:9], color="orange", label = "Cnidaria")
plt.scatter(x[9:15], y[9:15], color="yellow", label="Porifera")
plt.scatter(x[15:17], y[15:17], color="green", label="Xenacoelomorpha")
plt.scatter(x[17:23], y[17:23], color="blue",label="Vertebrata")
plt.scatter(x[23:35], y[23:35], color="Fuchsia",label="Ecdysozoa")
plt.scatter(x[35:], y[35:], color="DarkOrchid",label="Spiralian")

plt.xlabel("GC Content of Non Repeat Elements (%)")
plt.ylabel("GC Content of Repeat Elements (%)")
plt.legend()
plt.title("GC Content of Repeating vs Non Repeating Segments")

popt, pcov = np.polyfit(x,y,deg=1,cov=True)
err = np.sqrt(np.diag(pcov))#sqrt of diagonal elements of pcov are error
print(err)
popt=np.round(popt,8)#rounds popt to three digits
m,b = popt
func = np.poly1d(popt)
plt.plot([0,100],[0,100],color='grey',ls='--')
plt.plot(x,func(x), color='black')
label = str(m)+'x+'+str(b) #Can use latex in python
plt.text(30, 5,label)
```
